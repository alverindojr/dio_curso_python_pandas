**Listas**

```python
#criando lista chamada animais
animais = [1,2,3]

#imprime a variaval
animais
```

```python
#alterando a lista animais
animais = ['cachorro', 'gato', 12345, 6.5]

#imprimindo a lista 
animais
```

```python
#imprimindo o primeiro elemento da lista
animais[0]
```

```python
#imprimindo o último elemento da lista
animais[-1]
```

```python
#substituindo o primeiro elemento da lista
animais[0] = 'papagaio'
```

```python
#imprimindo a lista atualizada
animais
```

```python
#removendo gato da lista
animais.remove("gato")
```

```python
animais
```

```python
#retornar o tamanho da lista
len(animais)
```

```python
#verificando se existe o argumento "gato" na lista animais
"gato" in animais
```

```python
#criando uma lista numerica
lista = [500,30,300,80,10]
```

```python
#verificando o maior numero da lista
max(lista)
```

```python
#retornado o menor numero da lista
min(lista)
```

```python
#adicionar um unico elemento na lista
animais.append("leão")
```

```python
animais
```

```python
#adicionar mais de um elemento ao mesmo tempo na lista
animais.extend(["cachorro", "piriquito", "cobra"])
```

```python
animais
```

```python
#contando quantas vezes aparece um argumento na lista
lista.count(30)
```

```python
#ordenando a lista em ordem crescente
lista.sort()
```

**Dicionário**

```python
#dicionario trabalha com a sintaxe de chave e valor
dc = {"maça":5, "banana": 10, "morango": 30, "abacate": 37}
```

```python
dc
```

```python
#retornar o valor da chave morango
dc['morango']
```

```python
#o dicionario permite alteração dos seus valores, acessando pelas chaves
dc['morango'] = 50
```

```python
#imprimindo o valor atualizado
dc['morango']
```

```python
#retornando todas as chaves do dicionario
dc.keys()
```

```python
#retornando todos os valores do dicionario
dc.values()
```

```python
#verificando se existe uma chave e caso não exista insere
dc.setdefault('maca', 10)
```

```python
#no exemplo anterior maça já existia, agora com limao
dc.setdefault('limao', 10)
```
