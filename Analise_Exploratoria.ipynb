```python
#lendo o arquivo excel
df = pd.read_excel("AdventureWorks.xlsx")
```

```python
#retorna os tipos de cada coluna
df.dtypes
```

```python
#retorna a receita total
df['Valor Venda'].sum()
```

```python
#retornar o custo total
custo = df['Custo Unitário'] * df['Quantidade']
custo.sum()
```

```python
#criar a coluna custo total
df['Custo_Total'] = df['Custo Unitário'] * df['Quantidade']
```

```python
df.sample(5)
```

```python
#criando uma coluna de lucro
df['Lucro'] = df['Valor Venda'] - df['Custo_Total']
df.sample(5)
```

```python
#arredondar a coluna lucro
df['Lucro'] = round(df['Lucro'],2) 
```

```python
df.sample(5)
```

```python
#retorna o lucro total de todas as vendas
df['Lucro'].sum()
```

```python
#criando coluna tempo de envio
df['Tempo_Envio'] = (df['Data Envio'] - df['Data Venda']).dt.days
```

```python
df.sample(3)
```

```python
df.dtypes
```

```python
#retornar a media do tempo de envio por marca
df.groupby('Marca')['Tempo_Envio'].mean()
```

```python
#verificando se temos valores ausentes
df.isnull().sum()
```

```python
#salvar o arquivo em csv
df.to_csv("df_vendas_novo.csv", index=False)
```
